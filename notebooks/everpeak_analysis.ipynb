```python
import pandas as pd
import numpy as np

# 1. CREACIÓN MANUAL DE DATOS (Para evitar errores de conexión)
data = {
    'payment_method': ['credit_card', 'cash', 'debit_card', 'cash', 'credit_card', 'debit_card', 'cash', 'credit_card'],
    'quantity': [25, 10, 30, 5, 50, 15, 22, 100]
}
df = pd.DataFrame(data)

# 2. Función de segmentación avanzada
def classify_payment(row):
    card = row['payment_method']
    qty = row['quantity']

    # Manejo de valores nulos o faltantes
    if pd.isna(card) or pd.isna(qty):
        return "Error en Datos"

    # --- Segmentación por Volumen y Método de Pago ---
    if qty > 22:
        if card == 'credit_card' or card == 'debit_card':
            return "card_high_volume"
        else:
            return "no_card_high_volume"
    else: # qty <= 22
        if card == 'credit_card' or card == 'debit_card':
            return "card_low_volume"
        else:
            return "no_card_low_volume"

# 3. Aplicar la función al dataframe y verificar resultados
df['payment_segment'] = df.apply(classify_payment, axis=1)

# Mostrar el conteo exacto de cada segmento
print(df['payment_segment'].value_counts())
```

```python
# Análisis de missingness por columna (Práctica Guiada 1)
missing_summary = pd.DataFrame({
    'Valores Faltantes': df.isnull().sum(),
    'Porcentaje (%)': (df.isnull().sum() / len(df)) * 100
})
print(missing_summary.sort_values(by='Valores Faltantes', ascending=False))
```
